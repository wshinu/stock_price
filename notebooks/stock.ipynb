```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, r2_score

# Assuming you have a DataFrame named 'ms' with the columns you mentioned

# Select the input features (X) and the target variable (y)
input_columns = ['1. open', '2. high', '3. low', '6. volume']
target_column = '4. close'
X = ms[input_columns].values
y = ms[target_column].values

# Reshape the input data to 3D for LSTM (samples, time steps, features)
n_features = len(input_columns)
X_reshaped = np.reshape(X, (X.shape[0], 1, n_features))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)

# Scale the data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train[:, 0, :])
X_test_scaled = scaler.transform(X_test[:, 0, :])

# Reshape the scaled data to match the LSTM input shape
X_train_reshaped = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, n_features))
X_test_reshaped = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, n_features))

# Create and train the LSTM model
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(1, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train_reshaped, y_train, epochs=100, batch_size=32, verbose=0)

# Save the trained model to a file
model_file = 'lstm_model.h5'
model.save(model_file)

# Make predictions on the test set
y_pred = model.predict(X_test_reshaped)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, r2_score

# Load the data
data = ms  
# the '4. close' column contains the target variable
target_column = 'Close'
data = data[[target_column]].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Split the data into training and test sets
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

# Create sequences of inputs and corresponding outputs
def create_sequences(data, sequence_length):
    X = []
    y = []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)

sequence_length = 10  # Adjust the sequence length as needed
X_train, y_train = create_sequences(train_data, sequence_length)
X_test, y_test = create_sequences(test_data, sequence_length)

# Build the LSTM model
model = Sequential()
model.add(LSTM(64, input_shape=(sequence_length, 1)))
model.add(Dropout(0.2))  # Adding dropout regularization
model.add(Dense(1))
model.compile(optimizer=Adam(), loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Save the trained model to a file
model_file = 'lstm_model.h5'
model.save(model_file)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Rescale the predictions and the true values
y_pred = scaler.inverse_transform(y_pred)
y_test = scaler.inverse_transform(y_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```

```python
import pandas as pd
import numpy as np

def check_null_nan_values(df):
    """
    Function to check for null and NaN values in a DataFrame.
    
    Args:
    - df: pandas DataFrame
    
    Returns:
    - A DataFrame with columns indicating the presence of null and NaN values in the input DataFrame.
    """
    # Check for null values
    null_counts = df.isnull().sum()

    # Check for NaN values
    nan_counts = df.isna().sum()

    # Combine null and NaN counts into a single DataFrame
    result_df = pd.DataFrame({
        'Column': null_counts.index,
        'Null Count': null_counts.values,
        'NaN Count': nan_counts.values
    })

    return result_df
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from alpha_vantage.timeseries import TimeSeries

def predict_stock_price(api_key):
    # Step 1: Ask for user input for the stock symbol
    symbol = input("Please enter the stock symbol you want to predict: ")

    # Step 2: Retrieve daily adjusted stock prices for the symbol
    ts = TimeSeries(key=api_key, output_format='pandas')
    data, _ = ts.get_daily_adjusted(symbol=symbol, outputsize='compact')

    # Step 3: Preprocess the data
    target_column = '4. close'
    df = data[[target_column]].values

    # Step 4: Normalize the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df)

    # Step 5: Create sequences of inputs
    sequence_length = 10
    input_data = []
    for i in range(len(scaled_data) - sequence_length):
        input_data.append(scaled_data[i:i+sequence_length])
    input_data = np.array(input_data)

    # Step 6: Load the trained LSTM model
    model = Sequential()
    model.add(LSTM(64, input_shape=(sequence_length, 1)))
    model.add(Dense(1))
    model.compile(optimizer=Adam(), loss='mean_squared_error')
    model.load_weights('lstm_model.h5')

    # Step 7: Predict the stock prices
    predicted_data = model.predict(input_data)

    # Step 8: Rescale the predicted prices
    predicted_prices = scaler.inverse_transform(predicted_data)

    # Step 9: Retrieve the latest data for prediction
    latest_data, _ = ts.get_intraday(symbol=symbol, interval='1min')
    latest_price = latest_data.iloc[-1]['4. close']

    # Step 10: Prepare the input data for prediction
    input_data = scaled_data[-sequence_length:]
    input_data[-1] = latest_price
    input_data = np.expand_dims(input_data, axis=0)

    # Step 11: Perform prediction using the trained model
    predicted_price = model.predict(input_data)
    predicted_price = scaler.inverse_transform(predicted_price)

    return predicted_price[0][0]

# Example usage:


predicted_price = predict_stock_price(api_key)
print("Predicted price for tomorrow:", predicted_price)
```
